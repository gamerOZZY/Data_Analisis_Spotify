# Prueba basica

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

CLIENT_ID ='lol'
CLIENT_SECRET = 'lol'

sp = spotipy.Spotify(
    auth_manager=SpotifyClientCredentials(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET
    )
)

resultado = sp.search(q="Daft Punk", type="track", limit=1)

print(resultado)
```
